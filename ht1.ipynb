```python
ww_not_continuous_features = ['quality', 'color']
rw_not_continuous_features = ['quality', 'color']
ww_continuous_features = [col for col in white_wine_df.columns if col not in ww_not_continuous_features]
rw_continuous_features = [col for col in red_wine_df.columns if col not in rw_not_continuous_features]
assert ww_continuous_features == rw_continuous_features # проверили, совпадают ли названия признаков
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

dataframes = [
    ('Белые вина', white_wine_df),
    ('Красные вина', red_wine_df)
]

for name, df in dataframes:
    plt.figure(figsize=(15, 20))
    for i, col in enumerate(df.columns):
      if col in ww_continuous_features:
        plt.subplot(4, 3, i+1)
        sns.histplot(df[col], kde=True, color='blue')
        plt.title(f'{name} - {col}')
        plt.xlabel(col)
        plt.ylabel('Количество')

    # Вычисляем среднее и стандартное отклонение для всех признаков
    stats_texts = []
    for col in df.columns:
      if col in ww_continuous_features:
        mean = df[col].mean()
        std = df[col].std()
        stats_texts.append(f'{col}: μ={mean:.2f}, σ={std:.2f}')
    stats_str = '\n'.join(stats_texts)

    plt.suptitle(f'{name}\n{stats_str}', fontsize=12)

    # Если признаков меньше 12, отключаем лишний график
    if len(df.columns) < 12:
        plt.subplot(4, 3, len(df.columns)+1)
        plt.axis('off')

    plt.tight_layout(rect=[0, 0, 1, 0.95])  # чтобы не накладывался заголовок
    plt.show()
```

На построенных гистограммах распредления значений видно, что для многих показателей значения смещены относительно среднего в меньшую сторону.
На некоторых графиках видны большие значения признаков (сульфаты, диоксид серы, сахара, лимонная кислота), которые, вероятно, могут быть выбросами.

```python
for name, df in dataframes:
    plt.figure(figsize=(15, 20))
    for i, col in enumerate(ww_continuous_features[:12]):
        plt.subplot(4, 3, i+1)
        sns.boxplot(x=df[col], color='lightgray')
        plt.title(f'{name} - {col}')
        plt.xlabel('')
    plt.tight_layout()
    plt.show()
```

После построения boxplot подтверждаются выбросы или аномальные значения по содержанию некоторых химических соединений, которые могут негативно влиять на дольнейший анализ данных и требуют обработки. Заметен большой разброс по показателю "хлориды" для всех видов вин, "остаточный сахар" - для красных вин.
